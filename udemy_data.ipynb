```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load and display the dataset
dataset = "/content/gdrive/MyDrive/Udemy data/udemydataset.csv"
dataframe = df = pd.read_csv(dataset)
df
```

```python
df.info()
```

```python
# Converting datatype from object to datatime64
df["published_timestamp"] = pd.to_datetime(df["published_timestamp"])
df
```

```python
# Summary for numerical columns
df.describe()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
# Calculate Revenue (Only for Paid Courses)
df["Revenue"] = df["price"] * df["num_subscribers"]
df.loc[df["is_paid"] == False, "Revenue"] = 0
```

```python
df["subject"].unique()
```

```python
# Courses by subjects
subject_counts = df["subject"].value_counts()
subject_counts.plot(kind="bar", color="purple")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.title("Count of Courses by Subject")
plt.show()
```

```python
df["level"].unique()
```

```python
# Courses by subjects
subject_counts = df["level"].value_counts()
subject_counts.plot(kind="bar", color="purple")
plt.xlabel("Level")
plt.ylabel("Count")
plt.title("Count of Courses by Level")
plt.show()
```

```python
# Plot histograms for numerical columns
df.select_dtypes(include=['number']).hist(bins=30,figsize=(15, 10), color="purple")

plt.show()
```

```python
#Plot Heatmap
corr_matrix = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Extract Year, Month, and Weekday for trend analysis
df["Year"] = df["published_timestamp"].dt.year
df["Month"] = df["published_timestamp"].dt.month_name()
df["Weekday"] = df["published_timestamp"].dt.day_name()
```

```python
# Analyzing top subscribed courses
df.sort_values(by="num_subscribers", ascending=False).head(10)
```

Free courses attract the most students, with enrollments reaching over 250,000.
Paid courses generally have lower enrollments, especially as prices increase.
Some paid courses ($100-$200) still gain high enrollments, suggesting that there's no strong linear relationship between price & enrollments.

```python
# Scatter plot of course price vs. number of subscribers
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="price", y="num_subscribers", alpha=0.5, color="purple")
plt.title("Course Price vs. Number of Subscribers")
plt.xlabel("Price (USD)")
plt.ylabel("Number of Subscribers")
plt.show()
```

Courses with fewer than 100 lectures tend to attract the highest number of subscribers, with some exceeding 250,000 enrollments.

```python
# Visualize the Number of Lectures vs. Number of scubscribers
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="num_lectures", y="num_subscribers", alpha=0.5, color="purple")
plt.title("Number of Lectures vs. Number of Subscribers")
plt.xlabel("Number of Lectures")
plt.ylabel("Number of Subscribers")
plt.show()
```

Web Development is the most popular subject on Udemy, with over 7.8 million enrollments, indicating a high demand for coding and software development courses. Tech-related subjects dominate Udemy, likely due to industry demand for programming skills, while business and finance courses maintain steady interest.

```python
# Analyzing total number of subscribers per subject and reset index
subject_popularity = df.groupby("subject")[["num_subscribers"]].sum().reset_index()
subject_popularity = subject_popularity.sort_values(by="num_subscribers", ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(data=subject_popularity, x="num_subscribers", y="subject", hue=None, legend=False, color="purple")
plt.title("Total Number of Students Enrolled per Subject")
plt.xlabel("Number of Subscribers")
plt.ylabel("Subject Category")
plt.show()
```

```python
# Scatter plot of number of reviews vs. number of subscribers
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="num_reviews", y="num_subscribers", alpha=0.5, color="purple")
plt.title("Number of Reviews vs. Number of Subscribers")
plt.xlabel("Number of Reviews")
plt.ylabel("Number of Subscribers")
plt.show()
```

```python
# Compute correlation between reviews and subscribers
correlation_reviews = df[['num_reviews', 'num_subscribers']].corr().iloc[0, 1]
print(f"Correlation between Number of Reviews and Subscribers: {round(correlation_reviews, 3)}")
```

```python
# Sort courses by number of reviews
most_reviewed_courses = df.sort_values(by="num_reviews", ascending=False)[["course_title", "num_reviews", "num_subscribers"]]
```

```python
# Aggregate enrollments by difficulty level
level_stats = df.groupby("level")[["num_subscribers"]].sum().reset_index()

# Sort in descending order
level_stats = level_stats.sort_values(by="num_subscribers", ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(data=level_stats, x="num_subscribers", y="level", color="purple")
plt.title("Total Enrollments by Course Difficulty Level")
plt.xlabel("Number of Subscribers")
plt.ylabel("Course Difficulty Level")
plt.xscale("log")
plt.show()
```

```python
# Aggregate average reviews per difficulty level
level_engagement = df.groupby("level")[["num_reviews"]].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(data=level_engagement, x="num_reviews", y="level", color="purple")
plt.title("Average Number of Reviews by Course Difficulty Level")
plt.xlabel("Average Number of Reviews")
plt.ylabel("Course Difficulty Level")
plt.show()
```

```python
# Revenue by year
revenue_per_year = df.groupby("Year")["Revenue"].sum().reset_index()

# Plot revenue per year
plt.figure(figsize=(10, 5))
plt.plot(revenue_per_year["Year"], revenue_per_year["Revenue"], marker="o", linestyle="-", color="purple")
plt.title("Revenue by Year")
plt.xlabel("Year")
plt.ylabel("Total Revenue ($)")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
